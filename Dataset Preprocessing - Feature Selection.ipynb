```python
import pandas as pd

# Load your dataset

# Function to convert 'GDP' column to float
def convert_gdp(value):
    if isinstance(value, str):
        # Remove '$', 'B', and commas, and convert to float (considering the values are in billions)
        return float(value.replace('$', '').replace('B', '').replace(',', '').strip()) # Convert to actual float value
    return value

# Convert 'GDP' to float
df['GDP '] = df['GDP '].apply(convert_gdp)
df['Export '] = df['Export '].apply(convert_gdp)
df['Import '] = df['Import '].apply(convert_gdp)
df['Urban Population'] = df['Urban Population'].apply(convert_gdp)
df['Population '] = df['Population '].apply(convert_gdp)

# Function to convert 'Export Growth(%GDP)' to float
def convert_export_growth(value):
    if isinstance(value, str):
        # Remove '%', commas, and convert to float
        return float(value.replace('%', '').replace(',', '').strip())
    return value

# Function to convert 'GDP Per Capita' to float
def convert_gdp_per_capita(value):
    if isinstance(value, str):
        # Remove '$', commas, and convert to float
        return float(value.replace('$', '').replace(',', '').strip())
    return value

# Convert 'GDP Per Capita' to float
df['GDP Per Capita'] = df['GDP Per Capita'].apply(convert_gdp_per_capita)

# Show the updated DataFrame
print("Updated DataFrame with Converted Columns:")
print(df.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Calculate the correlation with 'Literacy Rate(%)'
correlation_with_target = df.corr()['Literacy Rate(%)'].drop('Literacy Rate(%)')

# Print the correlation values
print("Correlations with Literacy Rate(%):")
print(correlation_with_target)

# Plotting the correlations as a bar chart
plt.figure(figsize=(10, 6))
correlation_with_target.sort_values().plot(kind='barh', color='skyblue')
plt.title("Correlation of Features with Literacy Rate(%)")
plt.xlabel("Correlation Coefficient")
plt.ylabel("Features")
plt.axvline(0, color='black', lw=0.8, linestyle='--')  # Add a vertical line at 0 for reference
plt.grid(axis='x', linestyle='--', linewidth=0.5)
plt.show()
```

```python
correlation_dict = correlation_with_target.to_dict()
print(correlation_dict)
```

```python
# Select columns with high absolute correlation (greater than a threshold)
threshold = 0.8  # You can adjust this threshold as needed
selected_features = [col for col, corr in correlation_dict.items() if abs(corr) > threshold]
selected_features.append('Literacy Rate(%)')
# Create a new DataFrame with the selected features
selected_features_df = df[selected_features]

# Show the new DataFrame
print("New Dataset with Selected Features:")
selected_features_df.head()
```
